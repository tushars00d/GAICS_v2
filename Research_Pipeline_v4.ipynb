### Phase 1: High-Performance Generative Pre-training (Tabular DDPM)
Using Mixed Precision (AMP) to generate synthetic samples and augment the real CIC-IDS-2018 dataset.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from training.train_ddpm import train_ddpm
import os
import pandas as pd

sns.set_theme(style="whitegrid")

# Fix: Pre-process the CSV to remove classes with only 1 member before training
data_path = "/content/GAICS_v2/data/cicids_subset.csv"
if os.path.exists(data_path):
    df = pd.read_csv(data_path)
    label_counts = df.iloc[:, -1].value_counts()
    to_keep = label_counts[label_counts >= 2].index
    df_filtered = df[df.iloc[:, -1].isin(to_keep)]
    df_filtered.to_csv(data_path, index=False)
    print(f"[*] Filtered dataset: Removed {len(df) - len(df_filtered)} samples from rare classes.")

# Save current working directory
original_cwd = os.getcwd()
# Change to the project root directory where data and configs are located
os.chdir("/content/GAICS_v2")

try:
    ddpm_model, ddpm_history = train_ddpm("configs/default.yaml")
finally:
    # Always change back to the original working directory
    os.chdir(original_cwd)

if 'ddpm_history' in locals():
    plt.figure(figsize=(10, 4))
    plt.plot(ddpm_history['loss'], label="DDPM MSE Loss", color='purple', linewidth=2)
    plt.title("Tabular DDPM Generative Pre-training Loss")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.show()
```

### Phase 3: Explainability via SHAP (Addressing Viva Q8)
Generating quantitative SHAP values to explain which features the Attention-IDS focuses on.

```python
import shap
import torch
import numpy as np
from data.dataset_loaders import load_dataset
import yaml

with open("configs/default.yaml", "r") as f:
    config = yaml.safe_load(f)

_, test_loader, _, _ = load_dataset(config)
sample_data, _ = next(iter(test_loader))
sample_data = sample_data[:100].to(next(ids_model.parameters()).device)

explainer = shap.GradientExplainer(ids_model, sample_data)
shap_values = explainer.shap_values(sample_data)

shap.summary_plot(shap_values, sample_data.cpu().numpy(), feature_names=[f"F_{i}" for i in range(sample_data.shape[1])])
```

### Phase 4: Rigorous Ablation & Latency Profiling
Profiles end-to-end inference latency in milliseconds and calculates robust multi-seed Ablation Statistics.

```python
from evaluation.benchmark import run_ablation_studies
import IPython.display as display
from sklearn.metrics import roc_auc_score
import evaluation.metrics
import os
import functools

# Robust Patching: Force the multi_class parameter using partial or a wrapper
# This ensures that whenever calculate_roc_auc is called, it uses 'ovr'
def patched_roc_auc(y_true, y_prob):
    try:
        return roc_auc_score(y_true, y_prob, multi_class='ovr')
    except ValueError:
        # Fallback for binary cases if they occur during ablation
        return roc_auc_score(y_true, y_prob)

# Apply the patch directly to the metrics module
evaluation.metrics.calculate_roc_auc = patched_roc_auc

# Ensure we are in the correct directory to find the 'data' folder
original_cwd = os.getcwd()
os.chdir("/content/GAICS_v2")

try:
    # Run the studies with the robust patch in place
    results = run_ablation_studies("configs/default.yaml")
finally:
    os.chdir(original_cwd)

print("\nVisualizing Confusion Matrices from the final seed:\n")
if os.path.exists("plots/cm_clean.png"):
    display.display(display.Image(filename="plots/cm_clean.png"))
if os.path.exists("plots/cm_purified.png"):
    display.display(display.Image(filename="plots/cm_purified.png"))
```
